```python
board = [[' ',' ',' '],
         [' ',' ',' '],
         [' ',' ',' ']]

acceptable_row = [0,1,2]
acceptable_column = [0,1,2]

def user_choice(acceptable_row, acceptable_column):
    row, column = 'wrong','choice'

    while row.isdigit()==False and column.isdigit()==False:
        row = input('Enter the row in range(0-2)')
        column = input('Enter the column in range(0-2)')

        if row.isdigit()==False and column.isdigit()==False:
            print('Sorry, its not a number. Enter the row or column in numerical form between range(0-2):')

        else:
            if row not in acceptable_row and column in acceptable_column:
                print('Sorry, its out of range. Enter the row or column in numerical form between range(0-2):')
    
    print(row, ' -- ', column)

print('Welcome to the tic-tac-toe game!!')
print('Below is the board and its position in rows(vertical positioning) and columns(horizontal positioning)')
for i in range(3):
    print(i,end=" -- ")
    for j in range(3):
        print(j,end=" ")
        print(board[i][j],end="|")
    print("")

print('\n\nLets start the Game!!')
for i in range(3):
    for j in range(3):
        print(board[i][j],end="|")
    print("")

user_choice(acceptable_row, acceptable_column)



```
